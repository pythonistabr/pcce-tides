```python
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('../data/tide_variance_on_currents.csv', delimiter=';')
```

```python
plt.rcParams['font.size'] = 15
```

```python
data
```

```python
depth = np.arange(5.5, 53.5,2.5)
```

```python
fig, ax = plt.subplots(nrows=1, figsize=(10,8))
ax.bar(depth +.75, data.cross_shore,label='cross-shore')
ax.bar(depth, data.long_shore,label='along shore')
plt.ylabel('Variability(%)')
plt.xlabel('Depth(m)')
plt.legend(loc='upper right')
plt.show()
```
